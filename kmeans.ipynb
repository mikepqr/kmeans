# K-means in Python

## 2-dimensional data

`demo()` generates sample data and then uses `kmeans.kmeans()` to run the k-means algorithm on it. By default, 4 clusters are generated. You can change this with the keyword `K` in the `demo` call. The `demo` call is restricted to searching for the same number of clusters as are used to generate the data. 

The data is then shown in a simple scatter plot, colour-coded by the clusters derived by the algorithm.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import kmeans
kmeans.demo()
```

Here is another demonstration of `kmeans.kmeans()`, this time with fewer clusters in the sample data than are searched for by `kmeans` (in that sense, it is over-fitting). The `sigma` keyword can be

```python
x = kmeans.randomdata(K=3, sigma=0.1)
centroids, c, lowestdistortion = kmeans.kmeans(x, K=4)
kmeans.plotkmeans(x, c, centroids)
```
