# K-means in Python

## 2-dimensional data

`kmeans.demo()` generates sample data and then uses `kmeans.kmeans()` to run the K-means algorithm on this data.

The data is then shown in a simple scatter plot, colour-coded by the clusters derived by the algorithm.

By default, the simulated data has four clusters. You can change this with the keyword `K` in the `demo` call. `demo` is restricted to searching for the same number of clusters as are used to generate the data. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import kmeans
kmeans.demo()
```

Here is another demonstration of `kmeans()`, this time with fewer clusters in the sample data than are searched for by `kmeans` (the algorithm is over-fitting). 

Note that the `sigma` keyword can be passed to `randomdata()` to control the size of the simulated clusters.

```python
x = kmeans.randomdata(K=3, sigma=0.1)
centroids, c, lowestdistortion = kmeans.kmeans(x, K=4)
kmeans.plotkmeans(x, c, centroids)
```

## Higher-dimensional data

`kmeans()` works with data of abritrarily high dimensionality. Simple visualization is possible with up to 3 dimensions.

```python
kmeans.demo(ndim=3)
```
